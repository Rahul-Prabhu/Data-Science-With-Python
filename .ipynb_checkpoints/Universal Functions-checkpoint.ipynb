```python
angle_in_degrees=np.array([0,30,45,60,90,180])
angle_in_radians=angle_in_degrees*np.pi/180
```

```python
angle_in_radians
```

```python
print('sine values')
sine_values=np.sin(angle_in_radians)
sine_values
```

```python
print('cosine values')
np.cos(angle_in_radians)
```

```python
print('tan values')
np.tan(angle_in_radians)
```

```python
np.arcsin(sine_values)
```
