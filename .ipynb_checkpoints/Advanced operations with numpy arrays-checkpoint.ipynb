# Splitting Numpy Arrays

```python
p=np.arange(12)
p
```

```python
print('Split the arrays into 4 equal sized arrays')
np.split(p,4)
```

```python
print("Split the array into sub arrays according to the positions indicated in the function")
np.split(p,[3,8])
```

```python
array=np.array([['New York','Atlanta','Russia','Washington','Mumbai','Dallas'],
               ['Apple','Capgemini','NTT Data','L & T','Infosys','JP Morgan']])
```

```python
x1,x2=np.hsplit(array,2)
```

```python
x1
```

```python
x2
```

```python
np.hsplit(array,6)           #Column wise split of the array
```

```python
np.vsplit(array,2)          #Row wise split of the array
```

# Indexing

```python
x=np.arange(14)**2
x
```

```python
mask_1=[3,5,7,9]
x[mask_1]                     #Sansi Indexing
```

```python
mask_2=np.array([[2,5],
       [2,9]])
x[mask_2]
```

```python
cities=np.array((['Mumbai','Auckland','Bangalore'],
               ['Lisbon','Mangalore','Istanbul'],
               ['Moscow','Delhi','Barcelona']))
cities
```

```python
rows=[[0,0],[2,2]]
columns=[[1,2],[1,1]]
cities[rows,columns]
```

```python
cities[rows,columns]='aaaa'
cities
```

```python
x[mask_1]=999
```

# Indexing Mask

```python
import pandas as pd
```

```python
countries=pd.read_csv('C:/countries/countries of the world.csv',decimal=',')
countries.head()
```

```python
birthrates=countries['Birthrate'].values
type(birthrates)
```

```python
plt.plot(birthrates)
plt.show()
birthrates.shape
```

```python
birthrates=birthrates[~np.isnan(birthrates)]             #Removes nan from the array
birthrates.shape
```

```python
np.median(birthrates)
```

```python
np.mean(birthrates)
```

```python
np.count_nonzero(birthrates[birthrates<20])
```

```python
np.sort(birthrates)[0:20]
```

```python
birthrates[birthrates>40]
```

```python
np.count_nonzero((birthrates>30) & (birthrates<40))
```

# Boolean Mask

```python
x=np.arange(12).reshape(3,4)
x
```

```python
index_bool=x>8
index_bool
```

```python
x[index_bool]
```

```python
print(x[x>8])
```

```python
x<7
```

```python
np.count_nonzero(x<7)
```

```python
np.sum(x<7)
```

```python
np.sum(x<7,axis=1)
```

```python
np.any(x>8)
```

```python
np.all(x>7)
```

```python
np.all(x<50)
```

```python
np.all(x<10,axis=1)
```

# Array Structures

```python
emp_names=['Rahul','Kasturi','Karan','Umesh','Chitra','Rishabh']
emp_ids=[1,2,3,4,5,6]
emp_scores=[75,88,90,76,98,100]
```

```python
emp_data=np.zeros(6,dtype={'names':('Name','ID','Scores'),
                            'formats':('U16','i8','f8')})                   #Structure
```

```python
emp_data
```

```python
emp_data['Name']=emp_names
emp_data['ID']=emp_ids
emp_data['Scores']=emp_scores
emp_data
```

```python
emp_data['Name']
```

```python
emp_data['ID']
```

```python
emp_data['Scores']
```

```python
emp_data[2]
```

```python
emp_data[-2]['Name']
```

```python
emp_data[-2]['Scores']
```

```python
emp_data[emp_data['Scores']>80]['Name']
```

```python
emp_data[emp_data['Scores']<80]['Name']
```
