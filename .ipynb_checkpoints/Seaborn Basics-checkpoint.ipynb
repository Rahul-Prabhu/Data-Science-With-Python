# Simple univariate Functions

```python
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pollution_data=pd.read_csv('C:/countries/test.csv')
pollution_data
```

```python
sns.distplot(pollution_data['temperature'])
plt.show()
```

```python
dist=sns.distplot(pollution_data['temperature'],color='Green',vertical=True)
dist.set_title('Pollution Statistics')
dist.set_xlabel('Temprature of Region')
dist.set_ylabel('Density of Region')
plt.show()
```

```python
dist=sns.distplot(pollution_data['temperature'],           #kde:kernel distribution estimate
                 kde_kws={'color':'B',                     #kws:it is used to take key value arguments of an attribute
                         'lw':2,
                         'label':'KDE'},
                 hist_kws={'histtype':'step',
                          'linewidth':2,
                          'alpha':0.7,
                          'color':'g',
                          'label':'HIST'})
dist.set_title('Pollution Statistics')
dist.set_xlabel('Temprature')
dist.set_ylabel('Density')
plt.show()
```

```python
dist=sns.distplot(pollution_data['temperature'],color='Green',kde=False)
dist.set_title('Pollution Statistics')
dist.set_xlabel('Temprature')
dist.set_ylabel('Density')
plt.show()
```

```python
hist=sns.distplot(pollution_data['temperature'],color='Green',bins=80,kde=False)      #bins are used to make the bars thick or thin
hist.set_title('Pollution Statistics')
hist.set_xlabel('Temprature')
hist.set_ylabel('Frequency')
plt.show()
```

```python
rug=sns.rugplot(pollution_data['temperature'],
               height=0.7,
               axis='x')
rug.set_title('Rug Plot')
rug.set_xlabel('Temperature')

plt.show()
```

```python
plt.subplots(figsize=(20,5))

dist=sns.distplot(pollution_data['temperature'],bins=85,rug=True)

dist.set_title('Histogram,kde and Rug Plot')
dist.set_xlabel('Temprature')
dist.set_ylabel('Density')
plt.show()
```

```python
plt.subplots(figsize=(20,5))

dist=sns.distplot(pollution_data['temperature'],hist=False,rug=True)

dist.set_title('Histogram and Rug Plot')
dist.set_xlabel('Temprature')
dist.set_ylabel('Density')
plt.show()
```

# Bivariate Functions

```python
sns.set(color_codes=True)

plot=sns.kdeplot(pollution_data['temperature'],
                 shade=True,
                 color='b')
plot.set_title('KDE Plot')
plot.set_xlabel('Temperature')
plt.show()
```

```python
plt.subplots(figsize=(17,7))

sns.kdeplot(pollution_data['temperature'],
           color='k')

sns.kdeplot(pollution_data['temperature'],
           bw=0.3,
           label='bw:0.3',
           color='b')
sns.kdeplot(pollution_data['temperature'],
           bw=8,
           label='bw:8',
           color='g')
sns.kdeplot(pollution_data['temperature'],
           bw=3,
           label='bw:3',
           color='y')
sns.kdeplot(pollution_data['temperature'],
           bw=80,
           label='bw:80',
           color='r')
```

```python
joint_plot = sns.jointplot('temperature', 'weather_type', data=pollution_data, xlim={0,60},ylim={0,3})
plt.show()
```
