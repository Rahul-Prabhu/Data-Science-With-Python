```python
emp=np.array([['TCS','Capgemini','Reliance','IBM','Microsoft','Seimens','NTT Data'],
                   [50000,90000,89000,789000,450000,670000,76000],
                   [0.7,0.9,0.45,1.00,87.5,78.5,89.7]])
```

```python
emp
```

```python
i=0
for item in emp:
    print('Row',i,':',item)
    i+=1
```

```python
emp.flatten()
```

```python
emp
```

```python
for data in emp.flatten(order='F'):                           #Fortran order styling means column wise representation of data
    print(data)
```

```python
for data in emp.flatten():                           #C style ordering means row wise representation of data which is default
    print(data)
```

```python
num=np.arange(16).reshape(4,4)
num
```

```python
for i in num:
    print(i)
```

```python
for i in np.nditer(num):                      #used to print the array element wise
    print(i)
```

```python
for i in np.nditer(num,order='F',flags=['external_loop']):         #used to print the array column wise
    print(i)
```

```python
for array in np.nditer(num,op_flags=['readwrite']):
    array[...]=array*array
    
```
