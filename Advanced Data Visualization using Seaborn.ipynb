# Searching For Patterns in a Dataset

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings 
warnings.filterwarnings('ignore')
```

```python
employee_data=pd.read_excel('C:/countries/Absenteeism_at_work.xls')
employee_data
```

```python
plt.subplots(figsize=(12,6))
sns.distplot(employee_data['Reason for absence'],kde=False,rug=True,bins=29)
plt.show()
```

```python
plt.subplots(figsize=(12,6))
sns.distplot(employee_data['Month of absence'],kde=False,rug=True,bins=12)
plt.show()
```

```python
sns.jointplot(x='Day of the week',y='Reason for absence',data=employee_data,height=6)
plt.show()
```

```python
pairplot=sns.pairplot(employee_data,height=4,aspect=1,x_vars=['Distance from Residence to Work','Age','Body mass index'],y_vars=['Absenteeism time in hours','Work load Average/day '],kind='reg',hue='Disciplinary failure')
pairplot.fig.subplots_adjust(wspace=0.5,hspace=0.1)
plt.show()
```

# Distributions Within Categories

```python
filename1='C:/it_dspydsdj_06_enus_code/datasets/mpg.csv'
```

```python
cars_data=pd.read_csv(filename1)
cars_data.head()
```

```python
plt.subplots(figsize=(12,6))
sns.boxplot(x='cylinders',y='horsepower',data=cars_data)
plt.show()
```

Vertical lines of the box plot are whiskers. The boxes themselves represent the interquartile range of data distribution. The median value is represented by the horizontal line of the box. The dot is the outlier.

```python
plt.subplots(figsize=(12,6))
sns.violinplot(x='cylinders',y='horsepower',data=cars_data)
plt.show()
```

```python
plt.subplots(figsize=(12,6))
sns.violinplot("cylinders","horsepower",data=cars_data,bw=3,scale='count')
plt.show()
```

```python
plt.subplots(figsize=(12,6))
sns.violinplot("cylinders","horsepower",data=cars_data,bw=0.5,scale='count',inner='stick',palette='rainbow')
plt.show()
```

```python
plt.subplots(figsize=(12,6))
sns.violinplot("cylinders","horsepower",data=cars_data,inner=None)
sns.swarmplot("cylinders","horsepower",data=cars_data,color='black',size=3)
plt.show()
```

```python
sns.set(style='whitegrid',color_codes=True)
```

```python
plt.subplots(figsize=(12,6))
sns.stripplot('cylinders','horsepower',data=cars_data,size=6)
plt.show()
```

```python
plt.subplots(figsize=(12,6))
sns.stripplot('cylinders','horsepower',data=cars_data,size=6,jitter=True)
plt.show()
```

```python
plt.subplots(figsize=(12,6))
sns.swarmplot('cylinders','horsepower',data=cars_data)
plt.show()
```

```python
plt.subplots(figsize=(12,6))
sns.swarmplot('cylinders','horsepower',data=cars_data,hue='origin')
plt.show()
```

```python
plt.subplots(figsize=(12,6))
sns.barplot('cylinders','horsepower',data=cars_data,hue='origin')
plt.show()
```

```python
plt.subplots(figsize=(12,6))
sns.countplot('origin',palette='Accent',data=cars_data)
plt.show()
```

```python
plt.subplots(figsize=(12,6))
sns.pointplot('cylinders','horsepower',data=cars_data)
plt.show()
```

The Line connects the mean of each category while the vertical line represent the central tendencies

```python
sns.catplot(x='cylinders',y='horsepower',data=cars_data,row='origin',kind='box',height=6,aspect=2)
plt.show()
```

# Datasubsets using facet grids

```python
filename2='C:/it_dspydsdj_06_enus_code/datasets/Zomato.csv'
```

```python
Zomato_data=pd.read_csv(filename2)
Zomato_data.head()
```

```python
Zomato_data.describe()
```

```python
facet=sns.FacetGrid(Zomato_data,col="Price range") #One plot for each value of price range
plt.show()
```

```python
facet=sns.FacetGrid(Zomato_data,col="Price range")
facet.map(plt.hist,"Aggregate rating")
plt.show()
```

```python
facet=sns.FacetGrid(Zomato_data,col="Price range")
facet.map(plt.scatter,"Aggregate rating","Votes",color='0.3')
plt.show()
```

```python
facet=sns.FacetGrid(Zomato_data,col="Price range",hue='Currency')
facet.map(plt.scatter,"Aggregate rating","Votes",alpha=0.5)
facet.add_legend()
plt.show()
```

```python
facet=sns.FacetGrid(Zomato_data,row='Has Online delivery',col="Price range",margin_titles=True)
facet.map(sns.regplot,"Aggregate rating","Votes",fit_reg=False)
plt.show()
```

```python
facet=sns.FacetGrid(Zomato_data,row='Has Online delivery',col="Price range",margin_titles=True,gridspec_kws=dict(width_ratios=[2,3,5,3]))
facet.map(sns.regplot,"Aggregate rating","Votes",fit_reg=False)
plt.show()
```

```python
facet=sns.FacetGrid(Zomato_data,col='Currency')
facet.map(sns.barplot,'Aggregate rating','Average Cost for two')
plt.ylim(0,2000)
plt.show()
```

```python
facet=sns.FacetGrid(Zomato_data,col='Currency',col_wrap=3,aspect=1.5)
facet.map(sns.barplot,'Aggregate rating','Average Cost for two',color='#826EDF')
plt.ylim(0,2000)
plt.show()
```

```python
facet=sns.FacetGrid(Zomato_data,col='Currency',col_wrap=3,aspect=1.5)
facet.map(sns.barplot,'Aggregate rating','Average Cost for two',color='#826EDF')
facet.fig.subplots_adjust(wspace=0.5,hspace=0.5)
facet.set_axis_labels('Rating','Cost for Two')
plt.ylim(0,2000)
plt.show()
```
