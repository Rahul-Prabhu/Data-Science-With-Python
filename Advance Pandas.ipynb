# Multiindex Data Frames

# Grouping and Aggregations with Multiindex DataFrames

```python
multi_data.groupby(level='shape').sum()
```

```python
multi_data.groupby(level='colour').sum()
```

```python
multi_data.groupby(level=['colour','shape']).sum()
```

```python
multi_data.groupby(level=['colour','shape']).get_group(('white' ,'rectangle'))
```

```python
multi_data.groupby([pd.Grouper(level='colour'),'shape']).get_group(('white' ,'rectangle'))
```

```python
multi_data.groupby([pd.Grouper(level=0),'colour']).get_group(('rectangle' ,'white'))
```

```python
for name,group in objects_data.groupby('colour'):
    print('name:',name)
    print('group:',group)
```

```python
multi_data.groupby(level=['colour','shape']).size()
```

```python
multi_data.groupby(level=['colour','shape']).describe()
```

```python
multi_data.groupby(level=['shape']).aggregate(np.sum)
```

```python
multi_data.groupby(level=['shape','colour']).aggregate(np.sum)
```

```python
multi_data.groupby(level=['shape'],as_index=False).agg([np.sum,np.mean,np.min,np.max])
```

```python
multi_data.groupby(level=['shape']).agg({'quantity':np.sum,'objects':np.max})
```

```python
multi_data.groupby(level=['shape']).transform(lambda x:x.mean())
```

```python
multi_data['average']=multi_data.groupby(level=['shape']).transform(lambda x:x.mean())
```

# Reindexing

```python
candy_bars=pd.DataFrame({'name':['Snickers','Dairy Milk','KitKat','Oh Henry!','Ferrero Rocher'],
                        'company':['Mars','Cadbury','Nestle','Nestle','Ferrero SpA'],
                        'cover':['brown','purple','red','yellow','gold']})
```

```python
candy_bars
```

```python
candy_reindexed=candy_bars.reindex(index=[0,1,3],columns=['name','company'])
candy_reindexed
```

```python
candy_reindexed=candy_bars.reindex(index=[0,1,3,5,6],columns=['name','company','rating'])
candy_reindexed
```

```python
df=pd.DataFrame({'name':['name1','name2','name3','name4','name5','name6','name7','name8'],
               'company':['A','B','C','D','E','F','G','H']},index=[-2,-1,0,1,2,3,4,5])
df
```

```python
candy_bars_reindexed_like=candy_bars.reindex_like(df)
candy_bars_reindexed_like
```

```python
candy_bars_reindexed_like.ffill()
```

```python
candy_bars_reindexed_like.bfill()
```
