# Iterate Over

```python
import pandas as pd
```

```python
data={
    'Company':['IBM','HP','L&T','TCS','CAPGEMINI','WIPRO'],
    'CEO':['XYZ','XYZ','Sanjay Mehta','Ratan Tata','XYZ','Thiorrie Daolaporte'],
    'HEADQUARTERS':['USA','USA','INDIA','INDIA','FRANCE','INDIA']
}
```

```python
tech_company_df=pd.DataFrame(data)
tech_company_df
```

```python
for row in tech_company_df.itertuples():
    print(row)    
```

```python
for row in tech_company_df.itertuples():
    print('Company: ',row.Company)
    print('Headquarters: ',row.HEADQUARTERS,'\n')
    
```

```python
for row_index,row in tech_company_df.iterrows():
    print('Row Index: ',row_index)
    print('Company Name: ',row[0])
    print('Company CEO: ',row[1],'\n')
```

```python
for key,value in tech_company_df.iteritems():
    print('Key is: ',key)
    print('Value is:',value,'\n')
```

# Merging and Joining

```python
cars=pd.DataFrame({
    'Car':['Audi','WagonR','Fiat','Mercedes','BMW','NANO','Ferrari','Honda City'],
    'Color':['Red','Black','White','Blue','Green','Yellow','Purple','Blue'],
    'CarID':[1,2,3,4,5,6,7,8]
})
cars
```

```python
more_cars=pd.DataFrame({
    'Car':['Alto','Etios','Volkswagon'],
    'Color':['Red','Blue','White'],
    'CarID':[12,14,15]
})
# more_cars.index=[8,9,10]
more_cars
```

```python
total_car=pd.concat([cars,more_cars],ignore_index=True)
total_car
```

# Applying Join operations on Data Frame

```python
car_orders=pd.DataFrame({
    'OrderID':[201,202,203,204,205,206,207,208],
    'Quantity':[4,2,8,9,3,5,10,20],
    'CarID':[7,6,5,8,4,1,3,2]
})
car_orders
```

```python
pd.concat([cars,car_orders])   #incompatible operations
```

```python
pd.concat([cars,car_orders],axis=1)     #CarID is duplicated
```

```python
pd.merge(cars,car_orders,on='CarID')
```

```python
pd.merge(total_car,car_orders,on='CarID')
```

```python
pd.merge(total_car,car_orders,on='CarID',how='left')  
```

```python
pd.merge(total_car,car_orders,on='CarID',how='right')    #Inner Join Operations
```

```python
pd.merge(total_car,car_orders,on='CarID',how='outer')
```

```python
pd.merge(total_car,car_orders,on='CarID',how='inner')
```
