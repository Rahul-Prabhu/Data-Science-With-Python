# Intro to Pandas

```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
required_cols=['Country','Region','Population','Coastline (coast/area ratio)','GDP ($ per capita)','Literacy (%)','Birthrate']
```

```python
filename1='C:/it_dspydsdj_06_enus_code/visualizations/datasets/countries_of_the_world.csv'
```

```python
countries=pd.read_csv(filename1,usecols=required_cols,decimal=',')   #decimal attribute is used to replace comma or any other delimeter with dot
countries
```

```python
countries.rename(columns={'GDP ($ per capita)':'per capita GDP',
                        'Coastline (coast/area ratio)':'CoastlineRatio',
                        'Literacy (%)':'LiteracyRate'},inplace=True)
```

```python
countries.dropna(inplace=True)
```

```python
countries
```

```python
countries.Region.unique()
```

```python
avg_by_region=countries.groupby(by='Region').mean()
avg_by_region
```

# Transforming data using user defined functions

```python
change_case=lambda x:x.upper()
```

```python
cars_data['make'].apply(change_case)
```

```python
cars_data['current_location'].unique()
```

```python
cars_data['Import']=cars_data['current_location'].map({'In Tanzania':'No','Available For Import':'Yes'})
cars_data.head(10)
```

```python
def drop_comma(x):
    return str(x).replace(',','')
```

```python
cars_data['price_value']=cars_data['price_value'].apply(drop_comma)
```

```python
cars_data
```

```python
cars_data['price_value']=pd.to_numeric(cars_data['price_value'])
```

```python
cars_data['price_value']
```

```python
def category(price):
    price=int(price)
    if price<=10000000:
        return 'cheap'
    elif price>10000000 and price<30000000:
        return 'mid-range'
    elif price>30000000 and price<70000000:
        return 'expensive'
    else:
        return 'luxury'
    
```

```python
cars_data.price_value.apply(category).head(10)
```

```python
cars_data['price_range']=cars_data.price_value.apply(category)
```

```python
cars_data
```
